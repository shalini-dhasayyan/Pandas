Merge

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
dframe=DataFrame({'key':['x','q','y','z','x','x'],'dataset':np.arange(6)})
```

```python
dframe
```

```python
dframe1=DataFrame({'key':['q','y','z'],'dataset1':[1,2,3]})
dframe1
```

```python
pd.merge(dframe,dframe1)
```

```python
pd.merge(dframe,dframe1,on='key')
```

```python
pd.merge(dframe,dframe1,on='key',how='left')
```

```python
pd.merge(dframe,dframe1,on='key',how='right')
```

```python
pd.merge(dframe,dframe1,on='key',how='outer')
```

```python
dframe4=DataFrame({'key1':['x','q','y'],'key2':['xh','qu','yy'],'lft_data':[20,60,70]})
```

```python
dframe4
```

```python
dframe5=DataFrame({'key1':['x','q','y','u'],'key2':['xh','qu','yy','ii'],'rht_data':[20,60,70,88]})
```

```python
dframe4
```

```python
pd.merge(dframe4,dframe5,on=['key1','key2'],how='outer')
```

```python
pd.merge(dframe4,dframe5,on='key1')
```

```python
pd.merge(dframe4,dframe5,on='key1',suffixes=('lefty','righty'))
```

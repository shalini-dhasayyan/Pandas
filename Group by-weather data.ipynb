Group by

```python
import pandas as pd
df = pd.read_csv("weather_by_cities.csv")
df
```

```python
df1 = df.groupby("city")
df1
```

```python
for city, data in df1:
    print("city:",city)
    print("\n")
    print("data:",data) 
```

```python
df1.get_group('mumbai')
```

```python
df1.count()
```

```python
df1.min()
```

```python
df1.max()
```

```python
# maximum temp in each of cities

df1 = df.groupby("city").max()
df1
```

```python
%matplotlib inline
df1.plot(kind='line')
```
